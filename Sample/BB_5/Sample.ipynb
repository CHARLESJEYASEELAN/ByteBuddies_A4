```python
import pandas as pd
import datetime

# Sample data
dates_1 = [datetime.date(2024, 1, 17), datetime.date(2024, 1, 11), datetime.date(2024, 1, 12),datetime.date(2024, 1, 16), datetime.date(2024, 1, 15), datetime.date(2024, 1, 14)]
points_1 = [1, 0, 1,1, 0, 1]

dates_3 = [datetime.date(2024, 1, 17), datetime.date(2024, 1, 11), datetime.date(2024, 1, 12),datetime.date(2024, 1, 16), datetime.date(2024, 1, 15), datetime.date(2024, 1, 14)]
points_3 = [11, 11, 7,11, 11, 7]

dates_2 = [datetime.date(2024, 1, 17), datetime.date(2024, 1, 11), datetime.date(2024, 1, 12),datetime.date(2024, 1, 16), datetime.date(2024, 1, 15), datetime.date(2024, 1, 14)]
points_2 = [2, 2, 1,2, 2, 1]

# Create DataFrames
df_1 = pd.DataFrame({'Dates': dates_1, 'Points_1': points_1})
df_2 = pd.DataFrame({'Dates': dates_2, 'Points_2': points_2})
df_3 = pd.DataFrame({'Dates': dates_3, 'Points_3': points_3})

# Merge DataFrames on 'Dates'
merged_df = pd.merge(df_1, df_2, on='Dates', how='outer').merge(df_3, on='Dates', how='outer')

# Fill NaN values with 0
merged_df = merged_df.fillna(0)

# Sum the points for each date
merged_df['Total_Points'] = merged_df['Points_1'] + merged_df['Points_2'] + merged_df['Points_3']

# Display the final DataFrame
final_Bar_Df = merged_df[['Dates','Total_Points']]
final_Bar_Df.head(5)
```

```python
import pandas as pd
from datetime import datetime

# Sample data
sentimentData = [
    {'_id': '65a7f68c86fa4239427e4300', 'positive': 0.592, 'negative': 0.0, 'neutral': 0.408, 'last_submission_three': datetime(2024, 1, 17, 15, 47, 24, 158000)},
    {'_id': '65a819d508b865b136e74e6a', 'positive': 0.592, 'negative': 0, 'neutral': 0.408, 'last_submission_three': datetime(2024, 1, 11, 15, 47, 24, 158000)},
    {'_id': '65a81a0b08b865b136e74e6d', 'positive': 0.592, 'negative': 0, 'neutral': 0.408, 'last_submission_three': datetime(2024, 1, 12, 15, 47, 24, 158000)},
    {'_id': '65a829eb08b865b136e74e70', 'positive': 0.792, 'negative': 0, 'neutral': 0.408, 'last_submission_three': datetime(2024, 1, 18, 15, 47, 24, 158000)}
]

# Convert the list of dictionaries to a DataFrame
sentimentDf = pd.DataFrame(sentimentData)

# Filter data for the present day
present_day_data = sentimentDf[sentimentDf['last_submission_three'].dt.date == datetime.now().date()]

# Extract positive, negative, and neutral scores for the present day
present_day_scores_df = present_day_data[['positive', 'negative', 'neutral']].transpose().reset_index()
present_day_scores_df.columns = ['Scores', 'Points']

# Display the DataFrame for present day

print(present_day_scores_df)


# Extract positive scores and dates
positive_scores_df = sentimentDf[['last_submission_three', 'positive']]
positive_scores_df.columns = ['Dates', 'Positive score']

positive_scores_df['Dates'] = positive_scores_df['Dates'].dt.date


```
